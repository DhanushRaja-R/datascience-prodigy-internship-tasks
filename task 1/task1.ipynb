```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the dataset is in a CSV file named 'population_data.csv'
df = pd.read_csv('population_data.csv', names=["Gender", "Age", "Country"])

# Print the first few rows of the dataset
print("Dataset:")
print(df.head())

# Calculate summary statistics for the Age column
print("\nSummary Statistics:")
print(df['Age'].describe())

# Calculate the mean age for each gender
print("\nMean Age by Gender:")
print(df.groupby('Gender')['Age'].mean())

# Plot a bar chart of the gender distribution
plt.figure(figsize=(8, 6))
df['Gender'].value_counts().plot(kind='bar')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()




```

```python
# Plot a histogram of the age distribution
plt.figure(figsize=(8, 6))
df['Age'].plot(kind='hist', bins=4)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot a histogram of the age distribution
plt.figure(figsize=(8, 6))
df['Age'].plot(kind='hist', bins=4)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python

# Plot a bar chart of the country distribution
plt.figure(figsize=(10, 8))
df['Country'].value_counts().plot(kind='bar')
plt.title('Distribution of Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

```python
# Plot a box plot of the age distribution by gender
plt.figure(figsize=(8, 6))
df.boxplot(column='Age', by='Gender')
plt.title('Age Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python

# Plot a box plot of the age distribution by country
plt.figure(figsize=(10, 8))
df.boxplot(column='Age', by='Country')
plt.title('Age Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Age')
plt.show()
```
