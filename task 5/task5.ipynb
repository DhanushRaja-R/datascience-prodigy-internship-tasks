```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv("accident.csv")
```

```python
# Check if 'Date' column exists
if 'Date' in df.columns:
    # Convert Date column to datetime format
    df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')
else:
    print("Error: 'Date' column not found in the dataset.")
```

```python
# Analyze the data
print("Traffic Accident Data Overview:")
print(df.head())
```

```python
# Plot the number of accidents by road condition and severity
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Road Condition', hue='Severity', palette='Set2')
plt.title('Accident Severity by Road Condition')
plt.ylabel('Number of Accidents')
plt.xlabel('Road Condition')
plt.legend(title='Severity')
plt.grid(True)
plt.show()
```

```python
# Plot the number of accidents by road condition and severity
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Road Condition', hue='Severity', palette='Set2')
plt.title('Accident Severity by Road Condition')
plt.ylabel('Number of Accidents')
plt.xlabel('Road Condition')
plt.legend(title='Severity')
plt.grid(True)
plt.show()
```

```python
# Plot the number of accidents by weather condition and severity
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Weather Condition', hue='Severity', palette='Set1')
plt.title('Accident Severity by Weather Condition')
plt.ylabel('Number of Accidents')
plt.xlabel('Weather Condition')
plt.legend(title='Severity')
plt.grid(True)
plt.show()
```

```python
# Scatter plot of accident hotspots (locations)
plt.figure(figsize=(10, 6))
plt.scatter(df['Longitude'], df['Latitude'], c=df['Severity'], cmap='coolwarm', s=100)
plt.colorbar(label='Severity')
plt.title('Accident Hotspots by Severity')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python
# Analyzing the impact of road conditions and weather
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Road Condition', hue='Weather Condition', palette='muted')
plt.title('Accidents by Road Condition and Weather Condition')
plt.ylabel('Number of Accidents')
plt.xlabel('Road Condition')
plt.legend(title='Weather Condition', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```
